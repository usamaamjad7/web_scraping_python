```python
from IPython.display import clear_output
import time
base_url = 'https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page='

owner = []
labels = []
price = []

for i in range(1,11):
  url = base_url+str(i)
  # we are pausing for 3 seconds
  time.sleep(3)
  print(f'Scraping info  from {i} page')
  clear_output(wait=True)

  response =  requests.get(url)
  soup = BeautifulSoup(response.text,'html')

 

  # owner information
  own = soup.find_all('span',attrs={'class': 'seller-type'})
  for o in own:
    owner.append(o.text)



  # sqrt_ft
  Lbl=soup2.find_all('td',attrs={'class': 'lbl rate'})
  
  for a  in Lbl:
    labels.append(a.text)

  #price
  prc = soup2.find_all('td',attrs={'class': 'price'})
  for p in prc:
    price.append(p.text)

```
